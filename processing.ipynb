```python
def find_bounding_rectangle(edges, image):
    contours, _ = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    largest_contour = max(contours, key=cv2.contourArea)
    x, y, w, h = cv2.boundingRect(largest_contour)
    annotated_image = image.copy()
    cv2.rectangle(annotated_image, (x, y), (x + w, y + h), (0, 255, 0), 2)
    return annotated_image, (x, y, w, h)

img = io.imread("test_2.png")
img =  rgb2gray(img[:, :, :3])
img = (img * 255).astype(np.uint8)

edges = canny(img, 2.5, 100, 150)

edges_uint8 = (edges * 255).astype(np.uint8)

rectangle_image, _ = find_bounding_rectangle(edges_uint8, img)
show_images([img, edges, rectangle_image])

img = io.imread("test_1.png")
img =  rgb2gray(img[:, :, :3])
img = (img * 255).astype(np.uint8)

edges = canny(img, 4, 10, 40)

edges_uint8 = (edges * 255).astype(np.uint8)

rectangle_image, _ = find_bounding_rectangle(edges_uint8, img)
show_images([img, edges, rectangle_image])

img = io.imread("test_3.png")
img =  rgb2gray(img[:, :, :3])
img = (img * 255).astype(np.uint8)

edges = canny(img, 2, 20, 70)

edges_uint8 = (edges * 255).astype(np.uint8)

rectangle_image, _ = find_bounding_rectangle(edges_uint8, img)
show_images([img, edges, rectangle_image])


img = io.imread("test_4.png")
img =  rgb2gray(img[:, :, :3])
img = (img * 255).astype(np.uint8)

edges = canny(img, 4, 20, 120)

edges_uint8 = (edges * 255).astype(np.uint8)

rectangle_image, _ = find_bounding_rectangle(edges_uint8, img)
show_images([img, edges, rectangle_image])
```
